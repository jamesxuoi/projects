### Visualising top frequent noun-phrases in Slack channel

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
from wordcloud import STOPWORDS
from textblob import TextBlob
```

```python
ds_mel['text'] = ds_mel['text'].str.lower()
ds_mel['text'].head()
```

```python
ds_mel['text'] = ds_mel['text'].str.replace('https','')
```

```python
def check_wordcloud(ds_mel, withall=True):
    if withall:
        text = ds_mel.text.values
    else:
        text = ','.join([x for x in ds_mel.text.values])
        blob = TextBlob(text)
        text = blob.noun_phrases()

    wordcloud = WordCloud(
        width=500,
        height=500,
        background_color='white',
        stopwords=STOPWORDS
    ).generate(str(text))

    plt.figure(figsize=(10, 10), facecolor='k', edgecolor='k')
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.show()
```

```python
check_wordcloud(ds_mel, withall=True)
```
